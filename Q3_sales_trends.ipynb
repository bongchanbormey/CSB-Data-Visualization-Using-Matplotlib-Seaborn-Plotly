```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Sales Data
data = {
    'Month': ['January', 'February', 'March', 'April', 'May'],
    'Product A Sales': [200, 220, 250, 270, 300],
    'Product B Sales': [150, 180, 210, 240, 280],
    'Product C Sales': [400, 390, 370, 360, 350]
}

df = pd.DataFrame(data)

# Set the Month as the index for easier plotting
df.set_index('Month', inplace=True)

# 1. Using Matplotlib: Line Chart
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Product A Sales'], marker='o', label='Product A')
plt.plot(df.index, df['Product B Sales'], marker='o', label='Product B')
plt.plot(df.index, df['Product C Sales'], marker='o', label='Product C')
plt.title('Sales Trends of Products A, B, and C Over the Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()


# 2. Using Seaborn: Bar Plot
df_melted = df.reset_index().melt(id_vars='Month', var_name='Product', value_name='Sales')
plt.figure(figsize=(10, 6))
sns.barplot(data=df_melted, 
            x='Month', 
            y='Sales', 
            hue='Product', 
            palette=['#FF9999', '#66B3FF', '#99FF99']) 

plt.title('Comparison of Product Sales by Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# 3. Using Plotly: Interactive Line Plot
fig = px.line(df_melted, 
            x='Month', 
            y='Sales', 
            color='Product', 
            markers=True, 
            title='Interactive Sales Trends of Products A, B, and C')
fig.update_layout(xaxis_title='Month', yaxis_title='Sales')
fig.show()
```
